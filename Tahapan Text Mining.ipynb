```python
import pandas as pd
import re
```

```python
#hapus komentar yang hanya berisi emoji

data = pd.read_csv(r'comment.csv')
df = data['comment']
comment = []
x = 0
y = ''
for i in df :
    emoji = i.replace('emoji','')
    if emoji.isspace(): 
        comment.append(x)
    else:
        y = 'tess'
    x += 1

comment_baru = data.drop(comment)
```

```python
komentar = comment_baru.drop(columns=['Unnamed: 0'])
komentar
```

```python
#hapus mention dan komentar yang hanya berisi mention
mention = []
for i in komentar['comment']:
    c=' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",i).split())
    d = c.replace('emoji','')
    mention.append(d)

komentar['comment'] = mention
komentar['comment'] = komentar[komentar['comment'] != '']
baru = komentar.dropna()
# baru = komentar.dropna()
# baru[200:220:]
```

```python
baru[:200]
```

```python
#melakukan case folding dan url dan angka dan whitespace()
case = []

for i in baru['comment']:
#     a = i.lower()
    b = re.sub(r'^https?:\/\/.*[\r\n]*', '', str(i), flags=re.MULTILINE)
    a = b.lower()
    c = re.sub(r'[0-9]+', '', a)
    d = c.strip()
    case.append(d)

baru['comment'] = case
baru
```

```python
baru.to_csv('hasil_pilah.csv', index=False)
```

```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for i in s:  
        str1 += str(i)   
    
    # return string   
    return str1  
```
